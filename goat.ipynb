# 45. Задача о пасущейся козе.

(1) Покажем, что данная задача сводится к решению следующего уравнения:

$\sin\beta - \beta\cos\beta = \frac{\pi}{2}$, где $\beta = 2\alpha$ (см. рисунок)


Пусть центр кругового поля находится в точке A. Ее радиус равен единице.  Веревка привязана к забору в точке B. Проведём окружность с радиусом r (длина веревки) из точки B. Пусть $\alpha - \angle EBC, \phi - \angle BAC, BC = DB = r, AF = x, FC = y.$ 

По условию задачи, площадь пересечения окружности с центром в точке A с окружностью с центром в точке B должна равняться $\frac{\pi}{2}$. Тогда, половина этой площади должна равняться $\frac{\pi}{4}$. Половину эту площади можно посчитать следующим образом: посчитать площадь сектора круга, ограниченного углом $\alpha$, плюс площадь сектора круга, ограниченного углом $\phi$, минус площадь треугольника ABC. Тогда, $\frac{\alpha  \pi  r^2}{2\pi} + \frac{\phi\pi1^2}{2\pi} - \frac{y *1}{2} = \frac{\pi}{4}$. Упростим выражение: $\alpha r^2 + \phi - y = \frac{\pi}{2}$. 

Теперь заметим, что $\angle ABC = \angle ACB = \frac{\pi - \phi}{2}$. Также верно, что $(1-x)^2 + y^2 = r^2$ и $x^2 + y^2 = 1$. Таким образом, $r^2 = 2 - 2x = 2(1-x)$.

Теперь,
$
\alpha r^2 + \phi - y = \frac{\pi}{2}
\\
\frac{\pi - \phi}{2} \cdot 2(1-x) + \phi - y = \frac{\pi}{2}
\\
\pi - \pi x - \phi + \phi x + \phi - y = \frac{\pi}{2}
\\
y - (\phi - \pi)x = \frac{\pi}{2}
\\
y + (\pi - \phi)x = \frac{\pi}{2}
$

Сделаем замену $x=\cos{\phi} , y = \sin{\phi}$, получим $\sin\phi + (\pi - \phi)\cos\phi = \frac{\pi}{2}$

Рассмотрим сумму углов четырехугольника BDAC:  
$4 \cdot\alpha + 2 \cdot \phi = 2\pi$ 

Отсюда, $2 \cdot \alpha = \pi - \phi;\space  \phi = 2 \cdot \alpha - \pi$, и тогда

$
\sin\phi + (\pi - \phi)\cos\phi = \frac{\pi}{2} \\
\sin{(2\alpha - \pi)} + 2\alpha\cos{(2\alpha - \pi)} = \frac{\pi}{2} 
\\
\sin{2\alpha} - 2\alpha\cos{2\alpha} = \frac{\pi}{2} 
\\
\sin\beta - \beta\cos\beta = \frac{\pi}{2}
$

(2) Теперь покажем, что $r = 2\cos\frac{\beta}{2}; \space \beta = 2\alpha$

По т. косинусов из 
$\triangle {ABC}$

$
r^2 = 1 + 1 - 2\cos{(\pi - 2\alpha)}
\\
r^2 = 2(1 - \cos{(\pi - \beta)}) \\
r^2 = 2(1 + \cos{\beta})  \\
\cos2a = 2\cos^2 a - 1 \\
1 + \cos2a = 2\cos^2 a \\
r^ 2 = 2(2\cos^2\frac{\beta}{2}) \\
r = 2\cos\frac{\beta}{2}
$

Построим график функции $r = 2\cos\frac{\alpha}{2}$

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(-20, 20, 200)
r = 2 * np.cos(x / 2)

plt.title("r = 2*cos(a / 2)")
plt.xlabel("alpha")
plt.ylabel("radius")

plt.plot(x,r, 'r')
plt.show()
```

Данное уравнение имеет бесконечное число решений.

Решим уравнение $\sin \alpha - \alpha \cos \alpha - \frac{\pi}{2} = 0$:

```python
from scipy import optimize

phi = lambda alpha: np.sin(alpha) - alpha * np.cos(alpha) - np.pi / 2 

beta = optimize.brentq(phi, 0, 4)
beta
```

Мы нашли одно из решений уравнения - $\alpha = 1.9$, теперь найдем длину веревки:

```python
length = 2 * np.cos(beta / 2)
length
```

Теперь нашли соответствующее значение длины веревки.

Вывод: хоть и уравнение $r = 2\cos(\frac{\alpha}{2})$ имеет бесконечное множество решений, для данной задачи подходит только одно - $\alpha = 1.9$ радиан, длина веревок же при этом будет равна приблизительно 1.15. Также стоит отметить, что простых нетрансцедентных уравнений для решения данной задачи нет (кроме интеграла от комплексных функций).
